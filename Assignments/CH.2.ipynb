```r
library(ISLR)
names(College)
head(College)

College[,0]
summary(College)

pairs(College[,1:10])
plot(College$Private, College$Outstate)
```

```r
Elite=rep('No', nrow(College))
Elite[College$Top10perc>50]="Yes"
Elite=as.factor(Elite)
dim(Elite)
```

```r
college=data.frame(College, Elite)
summary(college)

plot(college$Elite, college$Outstate)
```

```r
par(mfrow=c(2,2))
hist(college$P.Undergrad, breaks=50, xlim=c(0, 10000))
hist(college$F.Undergrad)
hist(college$S.F.Ratio)
hist(college$PhD)
```

```r
par(mfrow=c(1,2))
plot(college$Elite, college$PhD)
# most elite schools have a lot of phd
plot(college$F.Undergrad, college$Expend)
# the less the f.undergrad, the more the expend it seems
```
